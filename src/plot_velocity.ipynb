```python
import os
import matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import xarray as xr
import matplotlib as mpl
import seaborn as sns
from itertools import count
```

```python
project_path = '/Users/sebastian/Dropbox (MIT)/postdoc/'
sns.set(style='ticks',context='paper')
mpl.rc('figure', dpi=120, figsize=[10,5])
mpl.rc('savefig',dpi=500,bbox='tight')
mpl.rc('legend',frameon=False)
```

```python
data = xr.open_dataset(project_path+'data/xarray/xr_7780b_grid.nc')
```

```python
data = data.assign_coords(z=-data.z)
```

```python
data.time.min().values
```

```python
var = ['u1','v1']
f,ax=plt.subplots(len(var),1,sharex=True)
for i,ax in enumerate(ax):
    data[var[i]].plot(ax=ax,rasterized=True)
    ax.set(ylim=[-500,0],title=var[i],xlabel=None)
    ax.set_xticks(pd.date_range(data.time.min().values,periods=8,freq='M'))
# plt.subplots_adjust()
# plt.savefig('../figures/test.pdf')
```

```python
var = ['u','v1']
f,ax=plt.subplots(len(var),1,sharex=True)
for i,ax in enumerate(ax):
    data[var[i]].plot(ax=ax)
    ax.set(ylim=[-500,0],title=var[i],xlabel=None)
plt.subplots_adjust()
```
